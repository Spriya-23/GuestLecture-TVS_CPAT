```python
# Find the outliers using plot.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
auto_mg_data = pd.read_csv("/content/drive/MyDrive/TVS Guest Lecture/auto-mpg.csv")
Q1 = auto_mg_data['weight'].quantile(0.25)
Q3 = auto_mg_data['weight'].quantile(0.75)
IQR = Q3 - Q1

# Defining the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers
auto_mg_data['Outlier'] = ((auto_mg_data['weight'] < lower_bound) | (auto_mg_data['weight'] > upper_bound))

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(auto_mg_data['displacement'], auto_mg_data['weight'], c=np.where(auto_mg_data['Outlier'], 'red', 'blue'), label='Data Points')
plt.xlabel('displacement')
plt.ylabel('weight')
plt.title('Scatter Plot with Outliers Highlighted')
plt.legend(['Data Points', 'Outliers'])
plt.grid(True)
plt.show()
```

```python
"""Plot the correlation plot on the dataset and visualize giving an overview
    of relationships among data.
"""
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
auto_mg_data = pd.read_csv("/content/drive/MyDrive/TVS Guest Lecture/auto-mpg.csv")
# Calculating the correlation matrix
correlation_matrix = auto_mg_data[["weight","displacement","horsepower"]].corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
"""Fit a simple linear regression model using libraries such as Numpy or
      Scikitlearn.
               ( importLinearRegression from sklearn. linear_model)
 Import the packages and classes you need.
 Provide data for independent and dependent variables.
 Create a regression model and fit it with existing data.
 Check the results of model fitting to know whether the model is satisfactory.
"""
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
auto_mg_data = pd.read_csv("/content/drive/MyDrive/TVS Guest Lecture/auto-mpg.csv")

# Reshaping X for scikit-learn
X = auto_mg_data['weight'].values.reshape(-1, 1)  # Reshape to 2D array
Y = auto_mg_data['displacement'].values

# Creating and fitting the model
model = LinearRegression()
model.fit(X, Y)

# Getting the coefficient and intercept
coefficient = model.coef_[0]
intercept = model.intercept_

print(f"Coefficient (Slope): {coefficient}")
print(f"Intercept: {intercept}")

# Making predictions
Y_pred = model.predict(X)

# Plotting the data and the regression line
plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, Y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
import pandas as pd
from collections import Counter
import re
with open(file_path, 'r') as file:
    # Reading all lines from the file
    lines = file.read()



# Tokenize the text
tokens = re.findall(r'\b\w+\b', lines.lower())  # Convert text to lower case for consistency

# Count word frequencies
word_counts = Counter(tokens)

# Find the most frequent and least frequent words
most_frequent_word = word_counts.most_common(1)[0]  # (word, frequency)
least_frequent_word = min(word_counts.items(), key=lambda x: x[1])  # (word, frequency)

# Print results
print(f"Most frequent word: {most_frequent_word[0]} (Frequency: {most_frequent_word[1]})")
print(f"Least frequent word: {least_frequent_word[0]} (Frequency: {least_frequent_word[1]})")
```
